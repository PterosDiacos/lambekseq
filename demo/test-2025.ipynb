```python
from graphviz import Source
from lambekseq.semcomp import SemComp

SemComp.load_lexicon(
    abbr_path='../abbr.json', vocab_path='../schema.json'
)
```

```python
ex = 'patient `s mom `s mom had heart-disease'
pos = 'prp pos n pos n vt prp'
sc = SemComp(zip(ex.split(), pos.split()))
sc.unify()
```

```python
Source(sc.tokens[1].dot_styled)
```

```python
Source(sc.semantics[0].dot_styled)
```
